# Q3 Part 1

```python
import matplotlib.pyplot as plt
import cv2
from PIL import Image
import numpy as np
```

```python
def convolve(im, kernel):
    h, w = im.shape
    kernel = np.flip(kernel, 1)
    h_k, w_k = kernel.shape
    output = np.zeros(im.shape)
    im_padded = np.zeros((h + 2, w + 2))   
    im_padded[1:-1, 1:-1] = im
    
    for i in range(1,h): 
        for j in range(1,w):
            output[i,j]=(kernel*im_padded[i-1:i+2,j-1:j+2]).sum() 
            
    return output
```

```python
im = cv2.imread("box.png", 0)
kernel = np.array( [[ -1, -1, -1 ],[ 0, 0, 0], [1, 1, 1 ]] )

f, axs = plt.subplots(1,2,figsize=(15,15))
plt.subplot(121)
plt.imshow(im, cmap='gray')
plt.title("Original")

out = convolve(im, kernel)

plt.subplot(122)
plt.imshow(out, cmap='gray')
plt.title("After Convolution")
plt.show()
```

# Q3 Part2

```python
img = cv2.imread("blur.jpg", 0)
new = convolve(img, kernel)

kernel_t = np.transpose(kernel)
new_t = convolve(img, kernel_t)
cv2.normalize(new_t, new_t, 0, 255, cv2.NORM_MINMAX, dtype=-1)

f, axs = plt.subplots(figsize=(15,15))

plt.subplot(2,2,1), plt.imshow(img, cmap='gray'), plt.title("Original") 
plt.subplot(2,2,2), plt.imshow(new, cmap='gray'), plt.title("After Convolution")
plt.subplot(2,2,3), plt.imshow(new_t, cmap='gray'), plt.title("After Convolution with Transpose")

plt.show()
```

# Observations

In the first image, we are finding the edges along the vertical direction.
Whereas, in the second image, we find the edges along the horizontal direction.
