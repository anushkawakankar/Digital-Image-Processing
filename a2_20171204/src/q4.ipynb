# Q 4 Part 1

```python
import matplotlib.pyplot as plt
import cv2
import numpy as np
from math import floor
```

```python
def convolve(im, kernel):
    h, w, n = im.shape
    kernel = np.flip(kernel, 1)
    k, v = kernel.shape
    pad = floor(k/2)
    
    if(k%2==0):
        l = pad-1
    else:
        l = pad
            
    output = np.zeros(im.shape)
    im_padded = np.zeros((h + 2*pad, w + 2*pad, 3))  
    im_padded[pad:-pad, pad:-pad] = im
    for i in range(pad,h+pad): 
        for j in range(pad,w+pad):
            output[i-pad,j-pad, 0]=(kernel*im_padded[i-l:i+pad+1,j-l:j+pad+1, 0]).sum()
            output[i-pad,j-pad, 1]=(kernel*im_padded[i-l:i+pad+1,j-l:j+pad+1, 1]).sum()
            output[i-pad,j-pad, 2]=(kernel*im_padded[i-l:i+pad+1,j-l:j+pad+1, 2]).sum()
            
    return output
```

```python
def fil(k, a):
    fil = np.ones((k,k));
    fil = -1 * fil;
    fil[floor(k/2), floor(k/2)] = k*k*a - 1;
    
    return fil
```

```python
im = cv2.imread("bell.jpg")

out = convolve(im, fil(3, 2))
cv2.normalize(out, out, 0, 255, cv2.NORM_MINMAX, dtype=-1)
out = out.astype(np.uint8)

print("window size = 3x3, Weight factor= 2")

f, axs = plt.subplots(figsize=(15,15))
plt.subplot(121)
plt.imshow(cv2.cvtColor(im, cv2.COLOR_BGR2RGB))
plt.title("Original")
plt.subplot(122)
plt.imshow(cv2.cvtColor(out, cv2.COLOR_BGR2RGB))
plt.title("High Boost")
plt.show()
```

# Observations
