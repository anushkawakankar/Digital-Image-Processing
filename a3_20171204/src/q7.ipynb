```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

# Part 1

```python
def sample(im, nx, ny):
    h,w = im.shape
    new_im = []
    nx=round(h/nx)
    ny=round(w/ny)
    for i in range(0, h, nx):
        row = []
        for j in range(0, w, ny):
            row.append(im[i,j])
        new_im.append(row)
    return np.array(new_im).astype('uint8')
```

```python
im = cv2.imread('../input_data/bricks.jpg',0)

h, w = im.shape
plt.figure(figsize=(30,30))
plt.imshow(im, cmap='gray')
plt.title("Original Image",  fontsize=20)
for i in range(1,10):
    nx = h//i
    ny = w//i
    new_im = sample(im, nx, ny)       
    n = 330 + i
    plt.subplot(n)
    plt.title("nx = " + str(nx) + ", ny = " + str(ny),  fontsize=20)
    plt.imshow(new_im, cmap='gray')
plt.show()
```

```python
nx = h//2
ny = w//2
new_im = sample(im, nx, ny)
print("Nyquist Rate: ", nx/h, ",", ny/w)
plt.figure(figsize=(10,10))
plt.subplot(121)
plt.imshow(im, cmap='gray')
plt.subplot(122)
plt.imshow(new_im, cmap='gray')
plt.show()
```

# Part 2

```python
img = cv2.GaussianBlur(im, (15,15), 0)

h, w = img.shape
plt.figure(figsize=(30,30))
plt.subplot(431)
plt.imshow(img, cmap='gray')
plt.title("Blurred Image", fontsize=20)
for i in range(1,10):
    nx = h//i
    ny = w//i
    new_im = sample(img, nx, ny)       
    plt.subplot(4, 3, i+1)
    plt.title("nx = " + str(nx) + ", ny = " + str(ny),  fontsize=20)
    plt.imshow(new_im, cmap='gray')
    
plt.show()
```

```python
nx = h//3
ny = w//3
new_im = sample(img, nx, ny)
print("Nyquist Rate: ", nx/h, ",", ny/w)
plt.figure(figsize=(10,10))
plt.subplot(121)
plt.imshow(img, cmap='gray')
plt.subplot(122)
plt.imshow(new_im, cmap='gray')
plt.show()
```

The Nyquist rate decreases when the image is blurred
