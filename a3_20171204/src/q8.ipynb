```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
im = cv2.imread('../input_data/rectangle.jpg', 0)

h, w = im.shape
f = np.fft.fft2(im)
fshift = np.fft.fftshift(f)

M = cv2.getRotationMatrix2D((w/2, h/2), 40, 1.0)
rotated = cv2.warpAffine(im, M, (h, w)) 
f1 = np.fft.fft2(rotated)
fshift1 = np.fft.fftshift(f1)

plt.figure(figsize=(10, 10))
plt.subplot(221),plt.imshow(im, cmap = 'gray')
plt.title('Input Image')
plt.subplot(222),plt.imshow(abs(fshift), cmap = 'gray')
plt.title('FFT of rectangle.jpg')
plt.subplot(223),plt.imshow(rotated, cmap = 'gray')
plt.title('Rotated Image')
plt.subplot(224),plt.imshow(abs(fshift1) , cmap = 'gray')
plt.title('FFT of rotated rectangle.jpg')

plt.show()
```

```python
translation_matrix = np.float32([ [1,0,50], [0,1,50] ])
img_translation = cv2.warpAffine(im, translation_matrix, (w, h))
f1 = np.fft.fft2(img_translation)
fshift1 = np.fft.fftshift(f1)
magnitude_spectrum1 = 20*np.log(np.abs(fshift1))

plt.figure(figsize=(10, 10))
plt.subplot(121)
plt.imshow(img_translation, cmap='gray')
plt.subplot(122)
plt.imshow(abs(fshift1), cmap='gray')
plt.show()
```
