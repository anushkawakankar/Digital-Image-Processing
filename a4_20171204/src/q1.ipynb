```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def imflt(im,filt):
    out = im.copy()
    for j in range(1, im.shape[0] - 1):
        for k in range(1, im.shape[1] -1):
            if filterchck(im[j-1:j+2,k-1:k+2],filt):
                out[j,k] = 0
    return out
```

```python
def filterchck(im,filt):
    for j in range(im.shape[0]):
        for k in range(im.shape[1]):
            if filt[j,k]==-1:
                continue
            elif im[j,k]!=filt[j,k]:
                return False
    return True
```

```python
S1 = np.array([
            [0,0,0],
            [-1,1,-1],
            [1,1,1]])
S2 = np.array([
            [-1,0,0],
            [1,1,0],
            [-1,1,-1]])
```

```python
im = cv2.imread("../input_data/fingerprint.jpg", 0)

threshold = 1
binary = (im < 127).astype(np.uint8)


while True:
    out = binary
    for i in range(4):
        out = imflt(out,S1)    
        out = imflt(out,S2)    
        S1 = np.rot90(S1)
        S2 = np.rot90(S2)

    change = np.mean(np.abs(out-binary))
    if change <= threshold: 
        break
    else:
        binary = out
    
binary = (im < 127).astype(np.uint8)

plt.figure(figsize=(10,10))
plt.subplot(121)
plt.imshow(binary, cmap='gray')
plt.subplot(122)
plt.imshow(out, cmap='gray')
plt.show()
```

# 2 

S1 is used to detect edges and curves that are very similar to edges 

S2 is used to detect right angled triangles

# 3

1. We rotate the kernel by 90 degrees to detect horizontal and vertical edges.
2. It also helps in detecting right hand corners.
2. The skeleton becomes thinner with each iteration. That's why we use multiple passes.
